# Exploratory Data Analysis

### Suicide Rate (By Year)

```python
df1=suicidal.groupby(['Year','Gender'])['Total'].sum().reset_index().sort_values('Total',ascending=False)
plt.figure(figsize=(10,6))
sb.barplot(x='Year',y='Total',hue='Gender',data=df1,ci=False,palette='seismic')
```

The total number of suicides are been increasing fromm 2001 to 2011 but after this there's a slight decrease in suicide rate.

### Gender Suicide Frequency

```python
sb.barplot(x='Gender',y='Total',data=suicidal,ci=False)
```

### State having maximum number of Suicides.

```python
df2=suicidal.groupby(['State','Gender'])['Total'].sum().reset_index().sort_values('Total',ascending=False)
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sb.set_style('darkgrid')
sb.barplot(x='State',y='Total',hue='Gender',data=df2,ci=False,palette='magma_r')
```

Maharastra records maximum cases of Suicides in Males whereas West Bengal has highest Female suicide rate.

### Which Age group has committed maximum suicides.

```python
limit = suicidal[~(suicidal['Age_group']=='0-100+')]
```

```python
limit.head()
```

```python
limit.groupby(by='Age_group')['Total'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
sb.set_style('darkgrid')
sb.barplot(x='Age_group',y='Total',data=limit,ci=False,palette='Dark2')
```

From this we conclude that the Age group between 15-44 has highest suicide frequency, i.e., The Middle Age group. 
It's quite surprising that kids from 0-14 has a suicide rate which is scary.

### Which is the major cause of attempting suicides.

# WordCloud

```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
try:
    words = " ".join(suicidal['Type'])
    def punctuation_stop(text):
        """remove punctuation and stop words"""
        filtered = []
        stop_words = set(stopwords.words('english'))
        word_tokens = word_tokenize(text)
        for w in word_tokens:
            if w not in stop_words and w.isalpha():
                filtered.append(w.lower())
        return filtered


    words_filtered = punctuation_stop(words)
    text = " ".join([ele for ele in words_filtered])
    wc= WordCloud(random_state=24,stopwords=STOPWORDS, max_words = 2000, width =1600, height = 800)
    wc.generate(text)

    plt.figure(figsize=(20,10), facecolor='k')
    plt.imshow(wc,interpolation="bilinear")
    plt.axis('off')
    plt.show()
except Exception as e:
    print(e)
    pass
```
